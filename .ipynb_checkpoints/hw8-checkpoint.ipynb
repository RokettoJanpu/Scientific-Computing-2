```python
from distmesh import *
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# square with vertices [0,0] and [3,3],
# pfix: corners of square and some points on circle of radius 1 and center [1.5,1.5]

def fd(p):
    return drectangle(p,0,3,0,3)

pfix = [[0,0], [3,0], [3,3], [0,3]]
n = 100
t = 2*np.pi/n
angles = t*np.array(range(n))
pfix += [[np.cos(angles[j])+1.5, np.sin(angles[j])+1.5] for j in range(n)]
pts, tri = distmesh2D(fd, huniform, 0.1, [0, 3, 0, 3], pfix)
#plt.savefig("hw8 mesh.png")
```
