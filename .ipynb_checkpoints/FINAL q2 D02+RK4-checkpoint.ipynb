```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial condition
k0 = 10
s0 = 0.1
def u0_func(x):
    return (2*np.pi*s0**2)**(-1/4)*np.exp(-x**2/(4*s0**2) + 1j*k0*x)

# exact sol
def u_ex(x,t):
    fac = s0**(1/2)*(2*np.pi)**(-1/4)*np.exp(-(s0*k0)**2)
    aux1 = (s0**2 + 0.5*1j*t)**(-1/2)
    aux2 = (4*s0**4*k0**2 - x**2 + 4*1j*s0**2*k0*x)/(4*(s0**2 + 0.5*1j*t))
    return fac*aux1*np.exp(aux2)

# RHS of PDE, discretized
def D02(u):
    m = np.size(u)
    F = np.zeros(m,dtype=np.complex_)
    for j in range(m):
        F[j] = u[(j+1)%m] + u[(j-1)%m] - 2*u[j]
        #print(F[j])
    #print("STOP")
    return 1j/(2*h**2)*F
```

```python
# parameters, mesh
k = 0.0001
T = 0.4
N = int(T/k)
xL = -20
xR = 20
for m in [4096,256]:
    h = (xR-xL)/(m-1)
    mesh = np.linspace(xL,xR,m)
    u = np.zeros(m,dtype=np.complex_)
    u = u0_func(mesh)

    # iterate u
    t_plot = T/5*np.array(range(6))
    total_probs = []
    for n in range(N+1):
        u_abs = np.abs(u)
        total_probs.append(np.trapz(u_abs**2,mesh))
        t = n*k
        # plot |u| at specific values of t
        if t in t_plot:
            plt.plot(mesh,u_abs,label='num')
            u_exact = [u_ex(x,t) for x in mesh]
            plt.plot(mesh,np.abs(u_exact),label='exact')            
            plt.legend()
            plt.xlabel('x')
            plt.ylabel('|u|')
            plt.title('|u| at t = {} for m = {} and D_0^2+RK4'.format(t,m))
            plt.savefig('FINAL u_abs t = {} m = {} D02+RK4.png'.format(t,m))
            plt.show()
        # RK4 time integrator
        k1 = D02(u)
        k2 = D02(u + 0.5*k*k1)
        k3 = D02(u + 0.5*k*k2)
        k4 = D02(u + k*k3)
        u = u + k/6*(k1 + 2*k2 + 2*k3 + k4)

    # plot total probability
    plt.plot(np.linspace(0,T,N+1),total_probs)
    plt.xlabel('t')
    plt.ylabel('total probability')
    plt.title('total probability vs t for m = {} and D_0^2+RK4'.format(m))
    ax = plt.gca()
    ax.set_ylim([0,2])
    plt.savefig('FINAL prob m = {} D02+RK4.png'.format(m))
    plt.show()
```
