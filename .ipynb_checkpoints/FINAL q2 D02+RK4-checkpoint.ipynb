```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# initial condition
k0 = 10
s0 = 0.1
def u0(x):
    return (2*np.pi*s0**2)**(-1/4)*np.exp(-x**2/(4*s0**2) + 1j*k0*x)

# exact sol
def u_ex(x,t):
    fac = s0**(1/2)*(2*np.pi)**(-1/4)*np.exp(-(s0*k0)**2)
    aux1 = (s0**2 + 0.5*1j*t)**(-1/2)
    aux2 = (4*s0**4*k0**2 - x**2 + 4*1j*s0**2*k0*x)/(4*(s0**2 + 0.5*1j*t))
    return fac*aux1*np.exp(aux2)

# RHS of PDE, discretized
def D02(u):
    m = np.size(u)
    p = np.array(range(m))
    F = np.zeros(m,dtype=np.complex_)
    for j in p:
        F[j] = u[(j+1)%m] + u[(j-1)%m] - 2*u[j]
        #print(F[j])
    #print("STOP")
    return 1j/(2*h**2)*F
```

```python
def iterate_sol(u,k,h,method):
    m = np.size(u)
    p = np.array(range(m))
    
    # D0^2+RK4
    if method == 0:
        k1 = D02(u)
        k2 = D02(u + 0.5*k*k1)
        k3 = D02(u + 0.5*k*k2)
        k4 = D02(u + k*k3)
        v = u + k/6*(k1 + 2*k2 + 2*k3 + k4)
        
    # modified equation
    #if method == 1:
    
        
    return v
      
```

```python
# parameters, mesh
xL = -20
xR = 20
m = 4096
h = (xR-xL)/(m-1)
mesh = np.linspace(xL,xR,m)
k = h
T = 0.4
N = int(T/k)
u = np.zeros(m,dtype=np.complex_)
u = u0(mesh)

# iterate u
t_plot = T/5*np.array(range(6))
total_probs = []
print(t_plot)
for n in range(N+1):
    u_abs = np.abs(u)
    #print(np.max(u_abs))
    total_probs.append(np.trapz(u_abs**2,mesh))
    #print(total_probs[-1])
    t = n*k
    if t in t_plot: # plot |u| at specific values of t
        #u_exact = [u_ex(x,t) for x in mesh]
        #plt.plot(mesh,np.abs(u_exact),label='exact')
        plt.plot(mesh,u_abs,label='num')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title('|u| at t = {} for method {}'.format(t,method))
        #plt.savefig('FINAL sol t = {} method {}.png'.format(int(t),method))
        plt.show()
    # RK4 time integrator
    k1 = D02(u)
    k2 = D02(u + 0.5*k*k1)
    k3 = D02(u + 0.5*k*k2)
    k4 = D02(u + k*k3)
    u = u + k/6*(k1 + 2*k2 + 2*k3 + k4)
```
